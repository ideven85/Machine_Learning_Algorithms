```python
import numpy as np
import pandas as pd
import tensorflow as tf
import keras as ks
```

```python
df = pd.read_csv("./d5fe9cfa-3637-4844-9c69-2ea989db999e-house-price-full-1-.csv")
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
df
```

```python
y = df.pop("price")
X = df
```

```python
y
```

```python
X = scaler.fit_transform(X)
y = np.log(y)
```

```python
model = ks.Sequential(
    [
        ks.layers.Dense(2, activation="sigmoid", input_shape=(X.shape[-1], 1)),
        ks.layers.Dense(1, activation="linear"),
    ]
)
model.summary()
```

```python
model.compile(optimizer=ks.optimizers.SGD(), loss="mean_squared_error")
```

```python
model.fit(X, y.values, epochs=10, batch_size=32)
```

```python
model.predict(X)[:, 0]
```

Solve the following system of linear equations using TensorFlow:

x + y + z + w = 13

2x + 3y − w = −1

−3x + 4y + z + 2w = 10

x + 2y − z + w = 1

Ans: x = 2, y = 0, z = 6 , w = 5

```python
import tensorflow as tf
```

```python
W = tf.Variable(
    [[1, 1, 1, 1], [2, 3, 0, -1], [-3, 4, 1, 2], [1, 2, -1, 1]], dtype=tf.float32
)
y = tf.Variable([[13], [-1], [10], [1]], dtype=tf.float32)
W = tf.linalg.inv(W)
```

```python
X = tf.cast(tf.round(tf.matmul(W, y)), dtype=tf.int32)
```
