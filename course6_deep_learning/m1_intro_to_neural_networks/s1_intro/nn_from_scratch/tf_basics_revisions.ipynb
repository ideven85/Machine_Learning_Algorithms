```python
def loss_fn_classify_binary(W, X, y, b):
    try:
        with tf.GradientTape() as tape:
            log_odds = tf.reduce_sum(W * X, axis=1) + b
            # tf.reduce_sum(W * X, axis=1) results in a (1, ) vector to which b is added

            probabilities = tf.sigmoid(log_odds)

            loss = tf.reduce_mean(
                -y * tf.math.log(probabilities)
                - (1.0 - y) * tf.math.log(1 - probabilities)
            )
        a, b = tape.gradient(loss, [W, b])
        return f"Output:{a.numpy()}, Bias:{b.numpy()}"
    except Exception as e:
        return "None type has no gradient"
```
