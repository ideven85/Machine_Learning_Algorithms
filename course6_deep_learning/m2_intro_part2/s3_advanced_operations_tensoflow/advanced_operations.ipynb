## Find the minima of y=x^2-4

```python
lr = tf.constant(0.1, dtype=tf.float32)
tolerance = 1.0
```

```python
# inital x
x = tf.Variable(tf.random.normal((1,)), name="Initial_Variable", dtype=tf.float32)
x
```

```python
y_prev = x**2 - 4
print(f"Inital y: {y_prev}")
```

```python
# loop of gradient descent
while True:
    print(f"y_prev,{y_prev}")
    with tf.GradientTape() as gradient_decent:
        y = x**2 - 4.0
    dy_dx = gradient_decent.gradient(y, x)
    y_new = x - lr * dy_dx
    delta = tf.abs((y_new - y_prev) * 100.0 / y_prev).numpy()[0]
    print(f" New y: {y_new}, delta: {delta}")
    if delta <= tolerance:
        print("Minimized reached, breaking")
        break
    else:
        y_prev = y_new
```

```python
dy_dx
```

```python
num_iterations = []
x_values = []
y_values = []
# loop of gradient descent
i = 0
while True:
    i += 1
    num_iterations.append(i)
    print(f"y_prev,{y_prev}")
    with tf.GradientTape() as gradient_decent:
        y = x**2 - 4.0
    dy_dx = gradient_decent.gradient(y, x)
    x.assign(x - lr * dy_dx)
    x_values.append(x.numpy()[0])
    y_values.append(y.numpy()[0])

    y_new = x**2 - 4
    print(f"{x.numpy()[0]},{y_new.numpy()[0]}")
    delta = tf.abs((y_new - y_prev) * 100.0 / y_prev).numpy()[0]
    print(f" delta: {delta}")
    if delta <= tolerance:
        print("Minimized reached, breaking")
        break
    else:
        y_prev = y_new
```
