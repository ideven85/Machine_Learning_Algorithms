```python
import numpy as np
import pandas as pd
import sklearn
from sklearn.model_selection import train_test_split
import tensorflow as tf
```

```python
import os

os.environ["KERAS_TELEMETRY"] = "0"
```

```python
df = pd.read_csv("data/house-price.csv")
```

```python
X = df.copy()
y = X.pop("price")
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_model = scaler.fit_transform(X)
y_model = np.log(y)
```

```python
df_scaled = pd.DataFrame(X_model)
```

```python
X_train, X_val, y_train, y_val = train_test_split(
    df_scaled, y_model, test_size=0.3, random_state=42
)
```

## Sequential Model TF

```python
model = tf.keras.Sequential()
model.add(layer=tf.keras.layers.Input(shape=(X_train.shape[-1],)))
model.add(layer=tf.keras.layers.Dense(units=10, activation="relu"))
# model.add(layer=tf.keras.layers.Dropout(.2))
model.add(layer=tf.keras.layers.Dense(units=10, activation="relu"))
model.add(layer=tf.keras.layers.Dense(units=5, activation="relu"))
model.add(layer=tf.keras.layers.Dense(units=1, activation="linear"))
```

```python
model.summary()
```

```python
model.compile(
    optimizer=tf.keras.optimizers.SGD(),
    loss=tf.keras.losses.MSE,
    metrics=["mse"],
)
```

```python
import tqdm

model.fit(
    X_train,
    y_train,
    epochs=10,
    batch_size=32,
    validation_data=(X_val, y_val),
    verbose=1,
)
```

```python
norm_layer = tf.keras.layers.BatchNormalization(name="norm_layer1")
```

```python
model2_normalized1 = tf.keras.Sequential(
    [
        tf.keras.Input(shape=(X_train.shape[-1],)),
        tf.keras.layers.Dense(units=2, activation="sigmoid"),
        norm_layer,
        tf.keras.layers.Dense(units=1, activation="linear"),
    ]
)
```

```python
model2_normalized1.compile(
    optimizer=tf.keras.optimizers.SGD(),
    loss=tf.keras.losses.MSE,
    metrics=["mse"],
)
```

```python
model2_normalized1.fit(
    X_train,
    y_train,
    epochs=10,
    batch_size=32,
    validation_data=(X_val, y_val),
    verbose=1,
)
```
