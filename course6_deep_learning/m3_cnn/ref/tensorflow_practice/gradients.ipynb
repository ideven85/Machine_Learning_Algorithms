```python
import numpy as np
import tensorflow as tf
from tensorflow import GradientTape
```

```python
x = tf.Variable(10.0)
with GradientTape() as tape:
    y = (
        tf.math.pow(3, (0.2 * x))
        + 5 * (tf.pow(x, 2))
        - tf.math.log(x)
        + tf.math.tanh(x)
    )
print(tape.gradient(y, x))
```

```python
x = tf.Variable(5.0)
# e = tf.math.exp
with GradientTape(persistent=True) as tape:
    t = tf.math.sqrt(x**0.3 + 0.1)
    p = tf.math.exp(4 * t)
    q = p**2 - 4 * p
print(tape.gradient(p, x))
print(tape.gradient(q, x))
```

Using TensorFlow, find the minima of the function: Instead of using the tolerance to decide convergence of the solution, use step gradient method.

number of iteration = 10; learning rate = 0.1

Equation 7x**2 - 4x + 10

Answer y = 9.428 at x = 0.285

```python
lr = tf.constant(0.1)
x = tf.Variable(tf.random.normal([1]), dtype=tf.float32)
for i in range(10):
    with GradientTape() as tape:
        y = 7 * x**2 - 4 * x + 10
    dydx = tape.gradient(y, x)
    x.assign(x - lr * dydx)
    y_new = 7 * x**2 - 4 * x + 10
print(y_new)
print(x)
```
