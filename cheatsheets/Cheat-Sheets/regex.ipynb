```python
import re
```

```python
word = "Hello I am Deven aged 24, brother is 30"
pattern = "\d+"
result = re.findall(pattern, word)
print(*result)
```

```python
WORD_RE = re.compile(r"\w+")
```

```python
re.findall(WORD_RE, word)
```

```python
import pandas as pd
```

```python
s = pd.Series(["a1a2 d4eve5n", "b1", "c2"], index=["A", "B", "C"])
```

```python
s.str.extractall(r"[a-e](\d)+")
```

```python
WORD_RE.findall(word)
```

```python
x = WORD_RE.finditer(word)

for l in x:
    w = l.group()
    print(w)
```
