```python
import numpy as np
import torch
```

```python
class Linear(torch.nn.Module):
    def __init__(self, inp, out):
        super(Linear, self).__init__()
        self.W = torch.randn(inp, out)
        self.bias = torch.zeros(out)

    def forward(self, x):
        return x @ self.W.T + self.bias
```

```python
linear = Linear(2, 2)
```

```python
print(linear)
```

```python
x_input = torch.tensor([[1, 2.0]])
y_hat = linear(x_input)
```
