### Parsing Sentences

#### Part-of-Speech Tagging: Assign parts of speech (POS) to each word.

```python
import nltk

nltk.download("averaged_perceptron_tagger")
nltk.download("punkt")

text = "The cat sat on the mat."
words = nltk.word_tokenize(text)
pos_tags = nltk.pos_tag(words)
print(pos_tags)
# Output: [('The', 'DT'), ('cat', 'NN'), ('sat', 'VBD'), ('on', 'IN'), ('the', 'DT'), ('mat', 'NN')]
```

#### Syntax Tree Construction: Build a syntax tree to represent the grammatical structure.

Building a syntax tree helps visualize and understand the hierarchical structure of a sentence. The tree represents the syntactic structure according to the grammar rules.

```python
from nltk import CFG, ChartParser

grammar = CFG.fromstring("""
    S -> NP VP
    NP -> DT JJ NN
    VP -> VBZ PP
    PP -> IN NP
    DT -> 'The' | 'the'
    JJ -> 'quick' | 'brown' | 'lazy'
    NN -> 'fox' | 'dog'
    VBZ -> 'jumps'
    IN -> 'over'
""")
parser = ChartParser(grammar)
sentence = "The quick brown fox jumps over the lazy dog"
tokens = nltk.word_tokenize(sentence)
for tree in parser.parse(tokens):
    tree.pretty_print()
```

#### Depedency Parsing

Identify the dependencies between words to understand the syntactic structure. Dependency parsing focuses on the relationships between words in a sentence. Each word is connected to a head word, and the nature of this connection is labeled.

```python
import spacy

nlp = spacy.load("en_core_web_sm")
doc = nlp("John threw the ball to Mary.")
for token in doc:
    print(f"{token.text} -> {token.head.text} ({token.dep_})")
# Output:
# John -> threw (nsubj)
# threw -> threw (ROOT)
# the -> ball (det)
# ball -> threw (dobj)
# to -> Mary (prep)
# Mary -> to (pobj)
```

#### Constituency Parsing

Group words into phrases (constituents) that function as a unit within the sentence.

```python
from nltk.parse import CoreNLPParser

parser = CoreNLPParser(url="http://localhost:9000")
parse = next(parser.raw_parse("John threw the ball to Mary."))
parse.pretty_print()
```
