## Text Normalization and Feature Extraction

Text normalization and feature extraction are crucial steps in natural language processing (NLP) and text analysis. They help transform raw text data into a format suitable for machine learning models and other analytical techniques.

#### Text Normalization

Text normalization is the process of transforming text into a standard format. It involves several steps to ensure consistency and remove noise from the data.

- Lowercasing: Convert all text to lowercase to ensure uniformity.
- Removing Punctuation: Remove all punctuation marks.
- Removing Stopwords: Remove common stopwords that do not add much meaning to the text.
- Lemmatization: Convert words to their base or dictionary form.
- Removing Special Characters/Patterns: Handle URLs, email addresses, and numbers.


```python
import re
import spacy
import nltk
from nltk.corpus import stopwords

# Download necessary NLTK data
nltk.download("stopwords")
nltk.download("punkt")
nltk.download("wordnet")
nltk.download("omw-1.4")

# Load Spacy model
nlp = spacy.load("en_core_web_sm")

# Sample text
text = """
Hi Customer Support,

The service has become better. Previously it was good. I hoped for better service, but it was worst. I am not good with it. I recently ordered a laptop from your website - http://www.hp.com, but I received a damaged product and am not happy. Could you please arrange for a replacement or refund my money? Additionally, I was charged twice for the same order. Please look into this matter urgently. Also I am running low on patient, in fact my patience has already ran its limit.

Thank you,
John Doe
"""


# Step 1: Lowercasing
def to_lowercase(text):
    return text.lower()


# Step 2: Removing Punctuation
def remove_punctuation(text):
    return re.sub(r"[^\w\s]", "", text)


# Step 3: Removing Stopwords
def remove_stopwords(text):
    stop_words = set(stopwords.words("english"))
    words = nltk.word_tokenize(text)
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return " ".join(filtered_words)


# Step 4: Lemmatization
def lemmatize_text(text):
    doc = nlp(text)
    lemmatized_text = " ".join([token.lemma_ for token in doc])
    return lemmatized_text


# Step 5: Removing Special Characters/Patterns
def remove_special_patterns(text):
    # Remove URLs
    text = re.sub(r"http\S+|www\S+|https\S+", "", text, flags=re.MULTILINE)
    # Remove email addresses
    text = re.sub(r"\S*@\S*\s?", "", text)
    # Remove numbers
    text = re.sub(r"\d+", "", text)
    return text


# Combine all steps into a single normalization function
def normalize_text(text):
    text = to_lowercase(text)
    text = remove_special_patterns(text)
    text = remove_punctuation(text)
    text = remove_stopwords(text)
    text = lemmatize_text(text)
    return text


# Normalize the sample text
normalized_text = normalize_text(text)
print(normalized_text)
```

### Challenges in Text Normalization

#### Handling Ambiguity
- Challenge: Words and phrases can have multiple meanings depending on the context.
- Example: "Bank" can refer to a financial institution or the side of a river.
- Impact: Incorrect interpretation can lead to errors in downstream tasks such as sentiment analysis or entity recognition.

#### Dealing with Informal Language
- Challenge: Text data from social media, chat messages, and other informal sources often contain slang, abbreviations, and non-standard grammar.
- Example: "gr8" for "great", "u" for "you", and "btw" for "by the way".
- Impact: Normalizing informal language to its standard form is difficult but necessary for accurate analysis.

#### Handling Different Variants of Words
- Challenge: Different variants of words, such as British and American English, need to be standardized.
- Example: "Colour" (British) vs. "Color" (American).
- Impact: Failure to normalize these variants can lead to inconsistencies in text processing.

#### Retaining Important Information
- Challenge: Removing stopwords, punctuation, and other elements can sometimes remove useful information.
- Example: In sentiment analysis, negations like "not" are important but may be removed if treated as stopwords.
- Impact: Over-normalization can lead to loss of critical information affecting the model's performance.

#### Handling Special Characters and Patterns
- Challenge: Text often contains URLs, email addresses, hashtags, mentions, and other special patterns.
- Example: "@username", "#hashtag", "http://example.com".
- Impact: Removing these elements can lead to loss of contextual information; retaining them requires careful handling to avoid introducing noise.

#### Dealing with Different Languages and Dialects
- Challenge: Text normalization needs to handle multilingual data and various dialects within a language.
- Example: Spanish has regional variations like "computadora" (Latin America) vs. "ordenador" (Spain).
- Impact: Normalization rules need to be language-specific and adaptable to different dialects.

#### Handling Named Entities
- Challenge: Normalizing named entities such as people, places, and organizations without losing their identity.
- Example: "International Business Machines" should remain as "IBM".
- Impact: Incorrect normalization can distort the meaning and reduce the accuracy of named entity recognition (NER) systems.

#### Preserving Semantic Meaning
- Challenge: Ensuring that the normalization process does not alter the semantic meaning of the text.
- Example: Lemmatizing "better" to "good" changes the meaning.
- Impact: Over-normalization can lead to misinterpretation of the text's intent.

#### Handling Code-Switching
- Challenge: Texts, especially from social media, often contain code-switching (mixing languages within a single sentence).
- Example: "I went to the tienda to buy some groceries."
- Impact: Requires sophisticated techniques to handle multiple languages within the same context.

#### Scalability and Performance
- Challenge: Normalizing large volumes of text data efficiently.
- Impact: Ensuring that the normalization process is scalable and performs well on big data is essential for practical applications.

#### Addressing the Challenges
To address these challenges, a combination of approaches can be used:

- Contextual Understanding: Using advanced NLP models like BERT or GPT-3 to understand the context and disambiguate meanings.
- Custom Dictionaries and Rules: Creating custom dictionaries for slang, abbreviations, and regional variants.
- Hybrid Approaches: Combining rule-based and machine learning techniques for better accuracy.
- Language-Specific Models: Employing models and tools tailored for specific languages and dialects.
- Preprocessing Pipelines: Designing flexible and modular preprocessing pipelines that can be adjusted based on the text's domain and context.

```python
import re
import spacy
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from typing import List

# Download necessary NLTK data
nltk.download("stopwords")
nltk.download("punkt")
nltk.download("wordnet")
nltk.download("omw-1.4")

# Load Spacy model
nlp = spacy.load("en_core_web_sm")


# Define the text preprocessing pipeline
class TextPreprocessor:
    def __init__(self):
        self.stop_words = set(stopwords.words("english"))

    def to_lowercase(self, text: str) -> str:
        return text.lower()

    def remove_punctuation(self, text: str) -> str:
        return re.sub(r"[^\w\s]", "", text)

    def remove_stopwords(self, text: str) -> str:
        words = word_tokenize(text)
        filtered_words = [word for word in words if word.lower() not in self.stop_words]
        return " ".join(filtered_words)

    def lemmatize_text(self, text: str) -> str:
        doc = nlp(text)
        lemmatized_text = " ".join([token.lemma_ for token in doc])
        return lemmatized_text

    def remove_special_patterns(self, text: str) -> str:
        # Remove URLs
        text = re.sub(r"http\S+|www\S+|https\S+", "", text, flags=re.MULTILINE)
        # Remove email addresses
        text = re.sub(r"\S*@\S*\s?", "", text)
        # Remove numbers
        text = re.sub(r"\d+", "", text)
        return text

    def normalize_text(self, text: str) -> str:
        text = self.to_lowercase(text)
        text = self.remove_special_patterns(text)
        text = self.remove_punctuation(text)
        text = self.remove_stopwords(text)
        text = self.lemmatize_text(text)
        return text


# Sample text data
texts = [
    """
    Hi Customer Support,

    I recently ordered a laptop from your website, but I received a damaged product. Could you please arrange for a replacement or refund my money? Additionally, I was charged twice for the same order. Please look into this matter urgently.

    Thank you,
    John Doe
    """,
    "Check out this link: http://example.com and email me at test@example.com.",
    "u r gr8! btw, r u coming 2 the party 2nite?",
]

# Create an instance of the TextPreprocessor
preprocessor = TextPreprocessor()

# Apply the preprocessing pipeline to each text
normalized_texts = [preprocessor.normalize_text(text) for text in texts]

# Print the results
for i, text in enumerate(normalized_texts):
    print(f"Original Text {i+1}:\n{texts[i]}\n")
    print(f"Normalized Text {i+1}:\n{text}\n{'-'*80}\n")
```

```python
import re
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer


def normalize_text(text):
    text = text.lower()  # Lowercasing
    text = re.sub(r"[^\w\s]", "", text)  # Removing punctuation
    tokens = word_tokenize(text)  # Tokenization
    stop_words = set(stopwords.words("english"))
    tokens = [word for word in tokens if word not in stop_words]  # Removing stop words
    lemmatizer = WordNetLemmatizer()
    normalized_tokens = [lemmatizer.lemmatize(word) for word in tokens]  # Lemmatization
    return " ".join(normalized_tokens)


text = "The quick brown fox jumps over the lazy dog."
normalized_text = normalize_text(text)
print(normalized_text)  # Output: "quick brown fox jump lazy dog"
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Sample text data
documents = ["The cat sat on the mat.", "The dog sat on the log."]

# Create the Bag of Words model
vectorizer = CountVectorizer()
bow = vectorizer.fit_transform(documents)

# Feature names and vectors
print(vectorizer.get_feature_names_out())
print(bow.toarray())
```

### Term Frequency-Inverse Document Frequency (TF-IDF)

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Sample text data
documents = ["The cat sat on the mat.", "The dog sat on the log."]

# Create the TF-IDF model
vectorizer = TfidfVectorizer()
tfidf = vectorizer.fit_transform(documents)

# Feature names and vectors
print(vectorizer.get_feature_names_out())
print(tfidf.toarray())
```

### Word Embeddings

```python
from gensim.models import Word2Vec

# Sample sentences
sentences = [
    ["the", "cat", "sat", "on", "the", "mat"],
    ["the", "dog", "sat", "on", "the", "log"],
]

# Train Word2Vec model
model = Word2Vec(sentences, vector_size=100, window=5, min_count=1, workers=4)

# Get word vector for 'cat'
vector = model.wv["cat"]
print(vector)
```

### N-Grams

Extracts contiguous sequences of n items (words or characters) from text.

#### Types:
- Unigrams: Single words.
- Bigrams: Pairs of consecutive words.
- Trigrams: Triples of consecutive words.


```python
# from sklearn.feature_extraction.text import CountVectorizer

# Sample text data
documents = ["The cat sat on the mat.", "The dog sat on the log."]

# Create N-gram model (bigrams)
vectorizer = CountVectorizer(ngram_range=(2, 2))
X = vectorizer.fit_transform(documents)

# Feature names and vectors
print(vectorizer.get_feature_names_out())
print(X.toarray())
```

### Feature Hashing (Hashing Vectorizer)
Maps features to indices in a fixed-size vector space using a hash function.

```python
from sklearn.feature_extraction.text import HashingVectorizer

# Sample text data
documents = ["The cat sat on the mat.", "The dog sat on the log."]

# Create the Hashing Vectorizer model
vectorizer = HashingVectorizer(n_features=10)
X = vectorizer.fit_transform(documents)

# Display vectors
print(X.toarray())
```

### TF-IDF with Character N-grams

Uses TF-IDF on character-level n-grams rather than word-level n-grams, which can be useful for language-agnostic applications or text with many misspellings.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Sample text data
documents = ["The cat sat on the mat.", "The dog sat on the log."]

# Create the TF-IDF model with character n-grams
vectorizer = TfidfVectorizer(analyzer="char", ngram_range=(2, 3))
X = vectorizer.fit_transform(documents)

# Feature names and vectors
print(vectorizer.get_feature_names_out())
print(X.toarray())
```

### Doc2Vec

Extends Word2Vec to learn document-level embeddings, capturing the semantics of entire documents.

```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument

# Sample text data
documents = ["The cat sat on the mat.", "The dog sat on the log."]

# Tag documents
tagged_data = [
    TaggedDocument(words=doc.split(), tags=[str(i)]) for i, doc in enumerate(documents)
]

# Train Doc2Vec model
model = Doc2Vec(tagged_data, vector_size=20, window=2, min_count=1, epochs=100)

# Get document vector
vector = model.infer_vector(["The", "cat", "sat", "on", "the", "mat"])
print(vector)
```

### TextRank for Keyword Extraction

Uses the TextRank algorithm, similar to PageRank, to score and extract important keywords from a text.

```python
import spacy
from collections import Counter
from spacy.lang.en.stop_words import STOP_WORDS

# Load Spacy model
nlp = spacy.load("en_core_web_sm")

# Sample text
text = "Text summarization is the process of distilling the most important information from a source."

# Apply Spacy
doc = nlp(text)

# Extract keywords using TextRank
words = [token.text for token in doc if not token.is_stop and not token.is_punct]
word_freq = Counter(words)
max_freq = max(word_freq.values())

for word in word_freq.keys():
    word_freq[word] = word_freq[word] / max_freq

print(word_freq)
```
