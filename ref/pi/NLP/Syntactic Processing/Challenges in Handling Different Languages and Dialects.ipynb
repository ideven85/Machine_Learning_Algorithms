```python
from gensim.models import Word2Vec

# Train a multilingual Word2Vec model
sentences = [["hello", "world"], ["bonjour", "le", "monde"], ["hola", "mundo"]]
model = Word2Vec(sentences, vector_size=100, window=5, min_count=1, workers=4)

# Access word vectors
vector_hello = model.wv["hello"]
vector_bonjour = model.wv["bonjour"]
```

## Morphological Analysis

Solution: Use language-specific stemmers and lemmatizers provided by libraries like NLTK and SpaCy.

```python
from nltk.stem import SnowballStemmer

# Stemming in English
stemmer = SnowballStemmer("english")
print(stemmer.stem("running"))  # Output: run

# Stemming in Finnish
stemmer = SnowballStemmer("finnish")
print(stemmer.stem("juokseminen"))  # Output: juoks
```
