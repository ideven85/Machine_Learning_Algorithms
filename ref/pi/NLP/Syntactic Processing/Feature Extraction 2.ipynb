#### Named Entity Recognition (NER)

Identifies and classifies named entities (e.g., people, organizations, locations) in text.

```python
import spacy

# Load pre-trained model
nlp = spacy.load("en_core_web_sm")

# Sample text
text = "Apple Inc. is looking at buying U.K. startup for $1 billion."

# Apply NER
doc = nlp(text)
for ent in doc.ents:
    print(ent.text, ent.label_)
```

#### Topic Modeling

Identifies topics in a collection of documents using probabilistic models.

##### Types:

- Latent Dirichlet Allocation (LDA): Assumes each document is a mixture of topics and each topic is a mixture of words.
- Non-Negative Matrix Factorization (NMF): Decomposes the document-term matrix into non-negative factors.


```python
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import CountVectorizer

# Sample text data
documents = [
    "The cat sat on the mat.",
    "The dog sat on the log.",
    "Cats and dogs are pets.",
]

# Create the Bag of Words model
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(documents)

# Apply LDA
lda = LatentDirichletAllocation(n_components=2, random_state=42)
lda.fit(X)

# Display topics
for idx, topic in enumerate(lda.components_):
    print(f"Topic {idx+1}:")
    print([vectorizer.get_feature_names_out()[i] for i in topic.argsort()[:-6:-1]])
```

#### Latent Semantic Analysis (LSA)

Reduces dimensionality by transforming the document-term matrix using singular value decomposition (SVD).

```python
from sklearn.decomposition import TruncatedSVD
from sklearn.feature_extraction.text import TfidfVectorizer

# Sample text data
documents = [
    "The cat sat on the mat.",
    "The dog sat on the log.",
    "Cats and dogs are pets.",
]

# Create the TF-IDF model
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(documents)

# Apply LSA
lsa = TruncatedSVD(n_components=2)
X_lsa = lsa.fit_transform(X)

# Display topics
print(X_lsa)
```

#### Text Summarization

Definition: Condenses a large text into a shorter version while preserving key information.

#### Types:

- Extractive Summarization: Extracts key sentences from the text.
- Abstractive Summarization: Generates new sentences that capture the essence of the text.


```python
from gensim.summarization import summarize

# Sample text
text = """Text summarization is the process of distilling the most important information from a source to produce an abridged version for a particular task and user (or user community). The aim is to create a coherent and fluent summary having only the main points outlined in the document. Summarization can be classified into extractive and abstractive summarization. Extractive methods select a subset of existing words, phrases, or sentences in the original text to form the summary. Abstractive methods generate new phrases and sentences that convey the most important information from the original text, possibly with rephrasing."""

# Summarize text
summary = summarize(text, ratio=0.4)
print(summary)
```

#### Sentence Embeddings
Represents entire sentences as dense vectors, capturing the semantics of the sentence.

##### Types:
- InferSent
- Universal Sentence Encoder (USE)
- BERT

```python
import tensorflow as tf
import tensorflow_hub as hub

# Load Universal Sentence Encoder
embed = hub.load("https://tfhub.dev/google/universal-sentence-encoder/4")

# Sample sentences
sentences = ["The cat sat on the mat.", "The dog sat on the log."]

# Generate embeddings
embeddings = embed(sentences)
print(embeddings)
```
