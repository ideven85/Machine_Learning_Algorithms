```python
# Reading the file using open()
file = open("Shakespeare_data.csv", mode="r", encoding="utf-8")

# Getting text data as string
text = file.read()

# Closing the file
file.close()
```

```python
lines = text.split("\n")
```

```python
for i in range(10):
    # print(lines[i])
    line = lines[i].split(",")
    print(line[5])
```

```python
import re

sentences = []
for line in lines:
    # get only the playerline
    l = line.split(",")

    if len(l) >= 5:
        l = l[5]

        # remove punctuation
        line = re.sub(r'[\!"#$%&\*+,-./:;<=>?@^_`()|~=]', "", l).strip()

        # tokenizer
        tokens = re.findall(r"\b\w+\b", line)

        if len(tokens) > 1:
            sentences.append(tokens)
```

```python
for i in range(10):
    print(sentences[i])
```

```python
# This is cbow implementation
# remember in Word2Vec we convert the text into a vector of real numbers.
# dimensionality is the length of the vector.
# More information - https://stackoverflow.com/questions/45394949/what-is-dimensionality-in-word-embeddings

# More information on below - https://radimrehurek.com/gensim/models/word2vec.html

from gensim.models import Word2Vec

shakes2Vec_cbow = Word2Vec(
    sentences,
    min_count=3,  # Ignore words that appear less than this
    size=50,  # Dimensionality of word embeddings - length of the the vectors
    sg=0,  # binary true or false - if sg is 0, it would pick up cbow,
    window=7,  # Context window for words during training
    iter=40,
)  # Number of epochs training over corpus
```

```python
shakes2Vec_cbow.wv.most_similar("husband")
```

```python
# This is skip gram implementation ------ sg  1

# remember in Word2Vec we convert the text into a vector of real numbers.
# dimensionality is the length of the vector.
# More information - https://stackoverflow.com/questions/45394949/what-is-dimensionality-in-word-embeddings

from gensim.models import Word2Vec

shakes2Vec_sg = Word2Vec(
    sentences,
    min_count=3,  # Ignore words that appear less than this
    size=50,  # Dimensionality of word embeddings - length of the the vectors
    sg=1,  # skipgrams
    window=7,  # Context window for words during training
    iter=40,
)  # Number of epochs training over corpus
```

```python
shakes2Vec_sg.wv.most_similar("husband")
```

#### How to evaluate?

There are two ways. 

#### Intrinsic evaluation 
involves comparing the embedding to a reference model. For example, comparing word similarity to an existing lexical database, such as WordNet. One can also manually annotate the embedding process results (not advisable, takes time and human resources, but if available great)

#### Extrinsic evaluation 
consists of evaluating the model's performance with regards to a downstream task such as text classification, machine translation, or summarization. The downstream task has its own performance evaluation strategy, which gives you insight into the embedding model's accuracy.
