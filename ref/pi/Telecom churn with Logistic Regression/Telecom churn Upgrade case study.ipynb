```python
from sklearn.experimental import enable_iterative_imputer  # noqa
from sklearn.impute import IterativeImputer
from sklearn.model_selection import train_test_split
import pandas as pd

# Example DataFrame with missing values
data = pd.DataFrame(
    {
        "feature1": [1, 2, None, 4, 5],
        "feature2": [None, 2, 3, 4, 5],
        "feature3": [1, 2, 3, None, 5],
    }
)

# Split data into training and test sets
train_data, test_data = train_test_split(data, test_size=0.2, random_state=42)

# Initialize the MICE imputer
mice_imputer = IterativeImputer(random_state=42)

# Fit the imputer on the training data and transform the training data
train_data_imputed = mice_imputer.fit_transform(train_data)

# Transform the test data using the same imputer (without refitting)
test_data_imputed = mice_imputer.transform(test_data)

# Convert the imputed arrays back to DataFrames
train_data_imputed = pd.DataFrame(train_data_imputed, columns=data.columns)
test_data_imputed = pd.DataFrame(test_data_imputed, columns=data.columns)

print("Imputed Training Data:")
print(train_data_imputed)

print("Imputed Test Data:")
print(test_data_imputed)
```
