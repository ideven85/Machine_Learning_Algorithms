```python
symbols = "abc#%^123ABC"
beyond_ascii = [s for s in symbols if ord(s) > 97]
```

```python
beyond_ascii
```

```python
# Comprehension using map_filter_reduce
beyond_ascii2 = list(filter(lambda c: c > 97, map(ord, symbols)))
```

```python
beyond_ascii2
```

```python
[str(x) for x in beyond_ascii2]
```
