```python
# Write code to display all the columns with their unique values and counts and check for redundant values
def unique_values_percentage_summary(df, columns):
    """
    A function to calculate the summary of categorical columns of a dataframe
        @param: df -> DataFrame
        @param: columns: list of categorical columns in the dataframe
        @returns: dictionary of summary of unique value counts and their percentages


    """
    results = {}
    for column in columns:
        # Count unique values and calculate percentages
        value_counts = df[column].value_counts()
        percentages = df[column].value_counts(normalize=True) * 100

        # Combine counts and percentages into a DataFrame
        unique_summary = pd.DataFrame(
            {"Count": value_counts, "Percentage (%)": percentages}
        )

        # Add the result to the dictionary
        results[column] = unique_summary

    return results
```

#### **4.4.1** Target likelihood analysis for categorical variables. <font color="red">[5 Marks]</font>
Investigate the relationships between categorical features and the target variable by analysing the target event likelihood (for the `'Y'` event) for each level of every relevant categorical feature. Through this analysis, identify categorical features that do not contribute much in explaining the variation in the target variable.

```python
# Write a function to calculate and analyse the target variable likelihood for categorical features


def level_wise_analysis(data, columns, target_variable):
    out = pd.DataFrame()
    for col in columns:
        df = data.copy()
        inner_df = df.groupby(col)[target_variable].mean()
        to_analyze = pd.DataFrame(
            {
                "Feature": col,
                "Category": inner_df.index,
                "Likelihood": np.round(inner_df.values, 2),
            }
        )
        out = pd.concat([out, to_analyze], ignore_index=True)
    # out=out.sort_values(by='Likelihood',ascending=False)
    return out
```

```python
df_train = pd.concat([X_train, y_train], axis=1)
df_train["fraud_reported"] = df_train["fraud_reported"].map({"Y": 1, "N": 0})
```

```python
categorical = categorical_columns.tolist()
categorical.remove("incident_hour_of_the_day")
```

```python
def target_likelihood_analysis(
    data, target, cat_vars, target_event="Y", cv_threshold=0.1
):

    results = []
    for var in cat_vars:
        grouped = (
            data.groupby(var)[target]
            .agg(
                Target_Likelihood=lambda x: (x == target_event).mean(),
                Sample_Size="count",
            )
            .reset_index()
            .rename(columns={var: "Category"})
        )
        grouped["Feature"] = var
        results.append(grouped)

    likelihood_df = pd.concat(results, ignore_index=True)

    cv_df = (
        likelihood_df.groupby("Feature")["Target_Likelihood"]
        .apply(
            lambda x: x.std() / x.mean() if x.mean() > 0 else 0
        )  # Avoid division by zero
        .reset_index()
        .rename(columns={"Target_Likelihood": "Coeff_of_Variation"})
    )

    # Step 3: Combine and add low contribution flag
    combined_table = likelihood_df.merge(cv_df, on="Feature")
    combined_table["Low_Contribution"] = (
        combined_table["Coeff_of_Variation"] < cv_threshold
    )

    # Format and sort
    combined_table["Coeff_of_Variation"] = combined_table["Coeff_of_Variation"].round(2)
    combined_table["Target_Likelihood"] = combined_table["Target_Likelihood"].round(2)
    combined_table = combined_table.sort_values("Coeff_of_Variation", ascending=False)

    # Reorder columns
    return combined_table[
        [
            "Feature",
            "Category",
            "Target_Likelihood",
            "Sample_Size",
            "Coeff_of_Variation",
            "Low_Contribution",
        ]
    ]
```

```python
analysis = level_wise_analysis(df_train, categorical, "fraud_reported")
analysis.sort_values(by="Likelihood", ascending=False)
```

```python
target_analysis = target_likelihood_analysis(
    df_train, "fraud_reported", categorical, 1, 0.1
)
target_analysis
```

#### **7.2.3** Evaluate VIF of features to assess multicollinearity <font color="red">[2 Marks]</font>

```python
# Import 'variance_inflation_factor'
from statsmodels.stats.outliers_influence import variance_inflation_factor


def variance_inflation(X, selected_features):
    vif = pd.DataFrame()
    vif["Features"] = X[selected_features].columns
    vif["VIF"] = [
        variance_inflation_factor(X[selected_features].values, i)
        for i in range(X[selected_features].shape[1])
    ]
    vif["VIF"] = round(vif["VIF"], 2)
    vif = vif.sort_values(by="VIF", ascending=False)
    return vif


variance_inflation(X_train_sm, X_train_sm.columns)
```

#### **7.3.1** Plot ROC Curve  to visualise the trade-off between true positive rate and false positive rate across different classification thresholds <font color="red">[2 Marks]</font>

```python
# Import libraries or function to plot the ROC curve
from sklearn.metrics import roc_curve, roc_auc_score


# Define ROC function
def draw_roc(actual, probs):
    fpr, tpr, thresholds = roc_curve(actual, probs, drop_intermediate=False)
    auc_score = roc_auc_score(actual, probs)
    plt.figure(figsize=(5, 5))
    plt.plot(fpr, tpr, label="ROC curve (area = %0.2f)" % auc_score)
    plt.plot([0, 1], [0, 1], "k--")
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel("False Positive Rate or [1 - True Negative Rate]")
    plt.ylabel("True Positive Rate")
    plt.title("Receiver operating characteristic example")
    plt.legend(loc="lower right")
    plt.show()

    return None
```

```python
def evaluate_model(dt_classifier, X_train, y_train, X_test, y_test):

    dt_classifier.fit(X_train, y_train)
    y_train_pred = dt_classifier.predict(X_train)
    y_test_pred = dt_classifier.predict(X_test)
    # print("Train Performance:", metrics.accuracy_score(y_train, y_train_pred))
    cm_train = metrics.confusion_matrix(y_train, y_train_pred)
    # print("-" * 50)
    # print(f"--Confusion matrix train: {cm_train} --\n\n")

    # print("Test Performance:", end=" ")
    # print(metrics.accuracy_score(y_test, y_test_pred))
    cm_test = metrics.confusion_matrix(y_test, y_test_pred)
    # print("-" * 50)
    # print(f"--Confusion matrix test: {cm_test} --")
    return (
        metrics.accuracy_score(y_train, y_train_pred),
        cm_train,
        metrics.accuracy_score(y_test, y_test_pred),
        cm_test,
    )
```
