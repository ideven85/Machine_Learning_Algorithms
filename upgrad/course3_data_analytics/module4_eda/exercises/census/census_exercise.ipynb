```python
# @np.vectorize(excluded={'0','p'})
def mypolyval(p, x):
    _p = list(p)
    res = _p.pop(0)
    while _p:
        res = res * x + _p.pop(0)
    return res
```

```python
# mypolyval([1, 2, 3], x=[0, 1])
```

```python
vpolyval = np.vectorize(mypolyval, excluded={0, "p"})  # excluded={0, 'p'}
vpolyval([1, 2, 3], x=[0, 1])
```

```python
vpolyval(p=[1, 2, 3], x=[1, 2])
```

```python
import scipy.stats

pearsonr = np.vectorize(scipy.stats.pearsonr, signature="(n),(n)->(),()")
pearsonr([[0, 1, 2, 3]], [[1, 2, 3, 4], [4, 3, 2, 1]])
```
