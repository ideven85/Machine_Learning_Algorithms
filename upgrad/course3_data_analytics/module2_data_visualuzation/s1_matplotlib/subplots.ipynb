````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
````

````python
years = np.array(["2012", "2013", "2014", "2015"])

sales_africa = np.array([127187.27, 144480.70, 229068.79, 283036.44])

sales_USCA = np.array([492756.60, 486629.30, 627634.98, 757108.13])

sales_LATAM = np.array([385098.15, 464733.29, 608140.77, 706632.93])

sales_Asia_Pacific = np.array([713658.22, 863983.97, 1092231.65, 1372784.40])

sales_Europe = np.array([540750.63, 717611.40, 848670.24, 1180303.95])
````

### Subplots with lineplots

We can use the following Matplotlib command to create subplots in Python:

```python
fig, ax = plt.subplots()
```

**It initiates a figure that will be used to comprise multiple graphs in a single chart.
Subplots are a good way to show multiple plots together for comparison. In the video above, you learnt how to plot different categories together in the same chart. Subplots offer the ability to create an array of plots, and you can create multiple charts next to each other to make it look like the elements of an array.** 


````python
fig, ax = plt.subplots()
(europe,) = ax.plot(years, sales_Europe)
europe.set_label("Europe")

(usca,) = ax.plot(years, sales_USCA)
usca.set_label("USCA")
usca.set_dashes([2, 2, 2, 2])

(asia_pacific,) = ax.plot(years, sales_Asia_Pacific)
asia_pacific.set_label("Asia Pacific")
asia_pacific.set_dashes([2, 2, 10, 2])

(africa,) = ax.plot(years, sales_africa)
africa.set_label("Africa")
africa.set_dashes([2, 2, 5, 2])  # 2 point break 5 point line

(latin_america,) = ax.plot(years, sales_LATAM)
latin_america.set_label("Latin America")
latin_america.set_dashes([2, 2, 5, 2])

plt.title(
    "Sales beween 2012 and 2015 of Difference regions\n\n",
    fontdict={"fontsize": 20, "fontweight": 5, "color": "brown"},
)
plt.xlabel("Years", fontdict={"fontsize": 12, "fontweight": 5, "color": "cyan"})
plt.ylabel("Sales", fontdict={"fontsize": 12, "fontweight": 5, "color": "cyan"})
# plt.legend()
plt.legend(bbox_to_anchor=(1.31, 0.4))
plt.savefig("Sales_Demo.png")
plt.show()
````

````python
fig, ax = plt.subplots(ncols=3, nrows=2, sharex=True, sharey=True)
(europe,) = ax[0][0].plot(years, sales_Europe, color="yellow")
europe.set_label("Europe")
ax[0][0].set_title(
    "Sales in Europe", fontdict={"fontsize": 12, "fontweight": 5, "color": "brown"}
)

usca = ax[0][1].bar(years, sales_USCA, color="cyan")
usca.set_label("USCA")
# usca.set_dashes([2,2,2,2])


africa = ax[0][2].scatter(years, sales_africa, color="brown")
africa.set_label("Africa")
# africa.set_dashes([2,2,5,2]) # 2 point break 5 point line

(latin_america,) = ax[1][0].plot(years, sales_LATAM)
latin_america.set_label("Latin America")
latin_america.set_dashes([2, 2, 5, 2])

(asia_pacific,) = ax[1][1].plot(years, sales_Asia_Pacific, color="green")
asia_pacific.set_label("Asia Pacific")
asia_pacific.set_dashes([2, 2, 10, 2])

# plt.title("Sales beween 2012 and 2015 of Difference regions\n\n",fontdict={"fontsize":20,"fontweight":5,"color":"brown"},loc='left')
# plt.xlabel("Years",fontdict={"fontsize":12,"fontweight":5,"color":"cyan"})
# plt.ylabel("Sales",fontdict={"fontsize":12,"fontweight":5,"color":"cyan"})
# plt.legend()
# plt.legend(bbox_to_anchor=(1.31,.4))
# plt.savefig("Sales_Demo.png")
fig.tight_layout()
# plt.title("")
# plt.show()
````

````python
fig, ax = plt.subplots(ncols=4, sharey=True)

(europe,) = ax[0].plot(years, sales_Europe, color="blue")
europe.set_label("Europe")
ax[0].set_title(
    "Sales in Europe\n", fontdict={"fontsize": 12, "fontweight": 5, "color": "brown"}
)
# europe.get_xlabel(rotation=90)


usca = ax[1].bar(years, sales_USCA, color="cyan")
usca.set_label("USCA")
ax[1].set_title(
    "Sales in USCA\n", fontdict={"fontsize": 12, "fontweight": 5, "color": "brown"}
)

africa = ax[2].scatter(years, sales_africa, color="brown")
africa.set_label("Africa")
ax[2].set_title(
    "Sales in Africa\n", fontdict={"fontsize": 12, "fontweight": 5, "color": "brown"}
)

asia = ax[3].bar(years, sales_Asia_Pacific, width=0.5, color="brown")
latam = ax[3].bar(years, sales_LATAM, width=0.5, color="royalblue")
ax[3].legend((asia, latam), ("Africa", "Latam"))
ax[3].set_title(
    "Sales in Asia and Latin America\n",
    fontdict={"fontsize": 12, "fontweight": 5, "color": "brown"},
)

fig.set_size_inches(20.5, 5.5, forward=True)

fig.tight_layout()
plt.savefig("Subplots.png")
````
