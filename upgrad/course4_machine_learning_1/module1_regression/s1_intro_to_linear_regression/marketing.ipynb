```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
marketing = pd.read_csv("Marketing.csv")
```

```python
marketing
```

```python
df = marketing.copy()
```

```python
lin_reg = LinearRegression()
```

```python
lin_reg.fit(
    np.array(df["Marketing Budget (X) (In lakhs)"]),
    np.array(df["Actual Sales(Y) (In crores)"]),
)
```

```python
X = np.array(df["Marketing Budget (X) (In lakhs)"]).reshape(17, -1)
```

```python
y = np.array(df["Actual Sales(Y) (In crores)"]).reshape(17, -1)
```

```python
X.shape
```

```python
y.shape
```

```python
lin_reg.fit(X, y)
```

```python
bias = lin_reg.intercept_[0]
```

```python
slope = lin_reg.coef_[0][0]
```
