## Multinomial and Bernoulli Naive Bayes

For understanding Multinomial and Bernoulli Naive Bayes, we will start with a small example and understand the end to end process. In another notebook, we will build a full-fledged email spam classifier.

To start with, let's take a few sentences and classify them in two different classes - *education* or *cinema*. Each sentence will represent one document. In real-world cases, a document be any piece of text such as an email, a news article, a book review, a tweet etc.
The analysis and the algorithm involved doesn’t depend on the type of document we use.

```python
import numpy as np
import pandas as pd
import sklearn
```

```python
df = pd.read_csv("example_train.csv")
```

```python
df
```

```python
df["Class_mapped"] = df["Class"].map({"cinema": 0, "education": 1})
```

```python
df
```

```python
df.drop("Class", axis=1, inplace=True)
```

```python
# Converting data frame into numpy array
train = df.values
X_train = train[:, 0]
y_train = train[:, 1]
```

```python
y_train.dtype
```

```python
# Converting into integer type
y_train = y_train.astype("int")
```

```python
X_train, y_train
```

```python
y_train.dtype
```

```python
from sklearn.feature_extraction.text import CountVectorizer

help(CountVectorizer)
```

```python
vectorizer = CountVectorizer()
```

```python
vectorizer.fit(X_train)
```

```python
dir(vectorizer)
```

```python
vectorizer.vocabulary_
```

## Removing Stop Words

```python
vec = CountVectorizer(stop_words="english")
```

```python
vec.fit(X_train)
```

```python
vec.vocabulary_
```

```python
print(
    f"Features: {vec.get_feature_names_out()}, Length: {len(vec.get_feature_names_out())}"
)
```

```python
X_transform = vec.transform(X_train)
str(X_transform)
```

```python
print(X_transform)
```

```python
X_transform.toarray()
```

```python
pd.DataFrame(X_transform.toarray(), columns=vec.get_feature_names_out())
```

```python
X_transform.shape
```

## Doing it on test set

```python
df_test = pd.read_csv("example_test.csv")
```

```python
df_test
```

```python
df_test["Class"] = df_test["Class"].map({"cinema": 0, "education": 1})
```

```python
test = df_test.values
X_test = test[:, 0]
y_test = test[:, 1]
print(X_test, y_test)
```

```python
X_test_transformed = vec.transform(X_test)
X_test_transformed
```

```python
print(X_test_transformed)
```

```python
# convert to non-sparse array
X_test = X_test_transformed.toarray()
X_test
```

Consider first 4 rows of the output: (0,2), (0,5), (0,7) and (0,11). It says that the first document (index 0) has 7th , 2nd , 5th and 11th 'word' present in the document, and that they appear only once in the document- indicated by the right hand column entry.


Similarly, consider the entry (4,4) (third from bottom). It says that the fifth document has the fifth word present twice. Indeed, the 5th word('good') appears twice in the 5th document.

## Building model using Multinomial Naive Bayes

```python
# Importing
from sklearn.naive_bayes import MultinomialNB

# Instantiation
mnb = MultinomialNB()

# Fitting

mnb.fit(X_transform, y_train)

# Prediction

mnb.predict(X_test)
```

```python
mnb.predict_proba(X_test)
```

```python
print("Probability of class cinema predicted:", mnb.predict_proba(X_test)[:, 0][0])
print("Probability of class education predicted:", mnb.predict_proba(X_test)[:, 1][0])
```

## Building model with Bernoulli

```python
import sklearn.naive_bayes as naive_bayes

##  bernoulli instantiation

bn = naive_bayes.BernoulliNB()

## fit

bn.fit(X_transform, y_train)

bn.predict_proba(X_test)
```
