## Dimensionality Reduction

```python
df = pd.read_clipboard()
```

```python
df
```

```python
df["X"] / df["Y"]
```

```python
df["X"].corr(df["Y"])
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=1)
X = df[["X", "Y"]]
pca.fit(X)
```

```python
pca.components_
```

```python
pca.explained_variance_
```

```python
pca.explained_variance_ratio_
```
