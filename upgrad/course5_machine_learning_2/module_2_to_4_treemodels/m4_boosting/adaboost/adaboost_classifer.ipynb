```python
X = df.drop("Attrition", axis=1)
y = df["Attrition"]
```

```python
## Splitting into categorical and numeric
df_cat = X.select_dtypes(include=["object"])
df_num = X.select_dtypes(include=["int64"])
```

```python
df_cat = pd.get_dummies(df_cat, dtype=int, drop_first=True)
```

```python
df_cat.head(10)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
df_num_scaled = scaler.fit_transform(df_num)
```

```python
df_num_scaled = pd.DataFrame(df_num_scaled, columns=df_num.columns, index=df_num.index)
```

```python
df_num_scaled
```

```python
y_mapped = y.apply(lambda x: 1 if x == "Yes" else 0)
```

```python
y_mapped
```

```python
help(AdaBoostClassifier)
```

```python
X_adaboost = pd.concat([df_num_scaled, df_cat], axis=1)
```

```python
clf = AdaBoostClassifier(n_estimators=100, random_state=0)
```

```python
clf.fit(X_adaboost, y_mapped)
```

```python
clf.score(X_adaboost, y_mapped)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_adaboost.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_adaboost, y_mapped, train_size=0.7, random_state=0, stratify=y_mapped
)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
clf = AdaBoostClassifier(n_estimators=100, random_state=0)
```

```python
clf.fit(X_train, y_train)
```

```python
y_test_pred = clf.predict(X_test)
```

```python
clf.score(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(y_test, y_test_pred)
```

```python
cm = metrics.confusion_matrix(y_test, y_test_pred)
cm
```
