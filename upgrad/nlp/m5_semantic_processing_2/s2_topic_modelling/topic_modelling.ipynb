```python
reviews = []
training_data = os.path.abspath("../../imdb/train/pos")
for f in os.listdir(training_data):
    if f.endswith(".txt"):
        with open(os.path.join(training_data, f), "r") as file:
            reviews.append(file.read())
```

```python
print("\n\n*****\n".join([x for x in reviews[:3]]))
```

```python
vect = TfidfVectorizer()
vect.fit_transform(reviews)
```

```python
X = vect.fit_transform(reviews)
Y = pd.DataFrame(X.toarray(), columns=vect.get_feature_names_out())
len(Y)
```

```python
Y.shape
```

```python
vect1 = TfidfVectorizer(stop_words="english")
X1 = vect1.fit_transform(reviews)
Y1 = pd.DataFrame(X.toarray(), columns=vect.get_feature_names_out())
Y1[:3]
```

```python
Y1.shape
```

```python
N_TOPICS = 5
nmf = NMF(n_components=N_TOPICS)
W = nmf.fit_transform(X1)  # Document-topic matrix
H = nmf.components_  # Topic-term matrix
```

```python
N_TOPICS = 15
nmf = NMF(n_components=N_TOPICS)
W = nmf.fit_transform(X1)  # Document-topic matrix
H = nmf.components_  # Topic-term matrix
```

```python
words = np.array(vect.get_feature_names_out())
topic_words = pd.DataFrame(
    np.zeros((N_TOPICS, 10)),
    index=[f"Topic {i + 1}" for i in range(N_TOPICS)],
    columns=[f"Word {i + 1}" for i in range(10)],
).astype(str)
for i in range(N_TOPICS):
    ix = H[i].argsort()[::-1][:10]
    topic_words.iloc[i] = words[ix]

topic_words
```

```python
words = np.array(vect.get_feature_names_out())
topic_words = pd.DataFrame(
    np.zeros((N_TOPICS, 10)),
    index=[f"Topic {i + 1}" for i in range(N_TOPICS)],
    columns=[f"Word {i + 1}" for i in range(10)],
).astype(str)
for i in range(N_TOPICS):
    ix = H[i].argsort()[::-1][:10]
    topic_words.iloc[i] = words[ix]

topic_words
```

```python
topic_mapping = {
    "Topic 3": "Fiction",
    "Topic 7": "War",
    "Topic 8": "Comedy",
    "Topic 12": "Book Adaptation",
    "Topic 13": "Horror",
    "Topic 15": "Martial Arts / Action",
}
```

```python
W = pd.DataFrame(W, columns=[f"Topic {i + 1}" for i in range(N_TOPICS)])
W["max_topic"] = W.apply(lambda x: topic_mapping.get(x.idxmax()), axis=1)
W[pd.notnull(W["max_topic"])].head(10)
```
