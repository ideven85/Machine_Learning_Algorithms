```python
W = tn.Variable(
    [[1, 1, 1, 1], [2, 3, 0, -1], [-3, 4, 1, 2], [1, 2, -1, 1]], dtype=tn.float32
)
y = tn.Variable([[13], [-1], [10], [1]], dtype=tn.float32)
W_inverse = tn.linalg.inv(W)
ans = tn.cast(tn.round(tn.matmul(W_inverse, y)), dtype=tn.int32)

print(ans)
```

```python
w1 = tn.Variable([[0.5, 1, -1.2]])
print(w1.shape)
X = tn.Variable([[0.2], [1], [0.5]])
ans = w1 @ X
ans = ans + 0.01
```

```python
ans = 1 / (1 + np.e**-ans)
```
