# Word Sense Disambiguation

```python
from nltk import wsd
```

```python
import copy

wn = wordnet
```

```python
wn.synsets("die")
```

```python
X = "The die is cast."
Y = "Roll the die to get a 6."
Z = "What is dead may never die."
```

```python
from IPython.display import display

help(wn.synsets)
```

```python
wn.synsets("die", pos="v")  # that is better
```

```python
for word in wn.synsets("die", pos="n"):
    print(word.definition())
```

```python
wsd.lesk(
    ["I", "went", "to", "the", "bank", "to", "deposit", "money", "."], "bank", "n"
).definition()
```

```python
wsd.lesk(X.split(), "die", "s").definition()
```

```python
from pprint import pprint

print(
    wsd.lesk(
        ["I", "went", "to", "the", "bank", "to", "deposit", "money", "."],
        "deposit",
        "v",
    ).definition()
)
print(
    wsd.lesk(
        ["I", "went", "to", "the", "bank", "to", "deposit", "money", "."],
        "deposit",
        "n",
    ).definition()
)
print(wsd.lesk(X.split(), "die", "v").definition())
print(wsd.lesk(X.split(), "die", "n").definition())
print(wsd.lesk(Y.split(), "die", "n").definition())
print("\nWhat is dead\n", wsd.lesk(Z.split(), "die", wn.NOUN).definition())
```
