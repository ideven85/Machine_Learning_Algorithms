```python
import re

x = ""
string = "       Python 3.0, released in 2008, was a major revision of the language that is not completely backward compatible and much Python 2 code does not run unmodified on Python 3. With Python 2's end-of-life, only Python 3.6.x[30] and_ later are supported, with older versions still supporting e.g. Windows 7 (and old installers not restricted to 64-bit Windows)."
a1 = (
    re.sub(r"[\d\W_]+", " ", string).strip().lower()
)  # \W_ _ for underscore for underscore?
a1
```

```python
nltk_stopwords = set(nltk_stopword)
```

```python
a1
```

```python
out = ""
for x in a1.split(" "):
    if x not in nltk_stopwords:
        out += x + " "
out
```

```python
" ".join(
    [
        x
        for x in re.sub(r"[\d\W_]+", " ", string).strip().lower().split()
        if x not in nltk_stopwords
    ]
)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(stop_words="english")
vectorizer
```

```python
vectorizer.fit(string.split())
```

```python
pprint(vectorizer.vocabulary_.items())
```
