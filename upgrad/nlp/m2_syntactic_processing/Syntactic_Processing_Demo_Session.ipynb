```python
sentence = "Rahul is playing football in the ground"
words = nltk.word_tokenize(sentence)
pos_tags = nltk.pos_tag(words)
print(pos_tags)
```

The output [('Rahul', 'NNP'), ('is', 'VBZ'), ('playing', 'VBG'), ('football', 'NN'), ('in', 'IN'), ('the', 'DT'), ('ground', 'NN')] is the result of part-of-speech (POS) tagging by NLTK, which assigns a grammatical category (tag) to each word in the sentence.

## Here's the meaning of each tagged word:

- 'Rahul' (NNP): Proper noun, singular (a specific name)
- 'is' (VBZ): Verb, 3rd person singular present (helping verb here)
- 'playing' (VBG): Verb, gerund or present participle (action happening now)
- 'football' (NN): Noun, singular or mass
- 'in' (IN): Preposition or subordinating conjunction
- 'the' (DT): Determiner (a marker before a noun)
- 'ground' (NN): Noun, singular or mass

This POS tagging shows the role of each word in the sentence like nouns, verbs, prepositions, etc. It helps in understanding the sentence structure and is useful in many NLP tasks like parsing, named entity recognition, or machine translation.





# Common parts of speech:

- Noun (person, place, thing, or idea):

Example: "The dog barked." (dog = noun)

- Pronoun (replaces a noun):

Example: "She runs fast." (she replaces a person's name)

- Verb (action or state):

Example: "They play soccer." (play = verb)

- Adjective (describes a noun):

Example: "The blue car is fast." (blue describes car)

- Adverb (describes a verb, adjective, or other adverb):

Example: "He runs quickly." (quickly describes runs)

- Preposition (shows relationship, usually location or time):

Example: "The book is on the table." (on = preposition)

- Conjunction (connects words or sentences):

Example: "I want tea and coffee." (and = conjunction)

- Interjection (expresses emotion or surprise):

Example: "Wow! That was amazing." (wow = interjection)

## What is a corpus?

A corpus is a large and structured collection of authentic texts (written or spoken) in a language, assembled to represent that language or a particular variety of it. Corpora are typically machine-readable and carefully sampled to be representative of real-world language use.

Linguists and language researchers use corpora to analyze language patterns, study grammar, semantics, and usage in natural contexts. The data in corpora enables empirical, data-driven study of language rather than relying solely on theoretical descriptions. Corpora can include diverse text types such as newspapers, conversations, literature, and even transcriptions of speech.

In summary, a corpus is:

- A body of naturally occurring language data (text or speech)
- Machine-readable and structured
- Representative of a language or language variety
- Used for linguistic description, research, and natural language processing applications


This approach allows studying language as it is actually used in the real world, providing insights into patterns, frequency, and variation of language features.

## Here are some well-known linguistic corpora along with brief descriptions:

- Brown Corpus

One of the first major English language corpora compiled in the 1960s. It contains about 1 million words sampled from a wide variety of text genres in American English circa 1961. It is foundational for modern corpus linguistics.

- Penn Treebank

A corpus annotated with syntactic or parse tree information for each sentence. It contains text from the Wall Street Journal. This corpus is widely used for training and evaluating syntactic parsers and POS taggers.

- British National Corpus (BNC)

A large corpus (~100 million words) representing both spoken and written British English from the late 20th century. It covers a variety of genres and styles, useful for broad English language study.

- LOB Corpus (Lancaster-Oslo/Bergen Corpus)

A British English counterpart to the Brown Corpus, composed using similar sampling methods but for British texts from the 1960s.

- Frown and Flob Corpora

More recent American (Frown) and British (Flob) English corpora from the 1990s, constructed in the same style as Brown and LOB respectively to support diachronic comparison.

- International Corpus of English (ICE)

A set of corpora representing different varieties of English worldwide, with both spoken and written components, compiled using the same principles to support cross-varietal comparison.

- WordNet

While not a traditional corpus, WordNet is a lexical database grouping English words into sets of synonyms (synsets) and recording semantic relations between them. It’s extensively used in NLP applications.



## Types of tenses

#### Present Tenses
- Simple Present

Habitual actions or general truths.
Example: She walks to school.

- Present Continuous

Actions happening right now or near future plans.
Example: She is walking to school.

- Present Perfect

Actions completed at an unspecified time or affecting the present.
Example: She has walked three miles today.

- Present Perfect Continuous

Actions started in the past and continuing now.
Example: She has been walking for an hour.

- Past Tenses

Simple Past
Completed actions at a specific time in the past.
Example: She walked to school yesterday.

- Past Continuous

Ongoing actions in the past interrupted or happening simultaneously.
Example: She was walking when it started to rain.

- Past Perfect

Actions completed before another past action.
Example: She had walked before it started to rain.

- Past Perfect Continuous

Duration of an action that was ongoing before another past action.
Example: She had been walking for an hour when it started to rain.

- Future Tenses

Simple Future
Actions that will happen.
Example: She will walk to school tomorrow.

- Future Continuous

Ongoing actions at a future time.
Example: She will be walking at 8 AM tomorrow.

- Future Perfect

Actions completed before a future time or event.
Example: She will have walked three miles by noon.

- Future Perfect Continuous

Duration of an action up to a future point.
Example: She will have been walking for an hour by noon.


## What is a gerund?

A gerund is a verb form that ends in -ing and functions as a noun in a sentence. Although it looks like a verb, a gerund acts like a noun, which means it can be used as the subject, object, or complement in a sentence.

## Key points about gerunds:
- They are formed by adding -ing to the base form of a verb.
- They name an activity, action, or concept (not a person or thing).
- Gerunds can take objects and be modified by adverbs, like verbs do, but their role is to act as a noun.

## Examples of gerunds in sentences:
- Swimming is fun. (Gerund as the subject)
- I enjoy reading. (Gerund as the direct object)
- She is interested in painting. (Gerund as the object of a preposition)

## Difference from present participle:
- A present participle also ends in -ing but functions as a verb (part of continuous tenses) or adjective.
- Example: The running water is cold. (Present participle as adjective)
- Example: She is running fast. (Present participle as verb)

In contrast, gerunds function as nouns representing actions or concepts.

This makes gerunds a blend of verb and noun features essential for expressing actions abstractly within sentences.

## Chunking


Chunking is the process of grouping words together into larger meaningful units or "chunks" instead of processing each word individually. This helps make language easier to understand, remember, and produce more fluently.

For example, in English, common chunks include fixed phrases or collocations such as "by the way," "at the end of the day," or "make a point." Instead of interpreting each word separately, these groups are processed as a single unit with a specific meaning. Another example is the phrase "How's it going?" which is a chunk often used in conversation and processed as a whole rather than as individual words.

Chunking helps learners of language to develop fluency by recognizing and using these multi-word units naturally during communication.

```python
import nltk
from nltk import pos_tag, word_tokenize
from nltk.chunk import RegexpParser
```

```python
sentence = "The quick brown fox jumps over the lazy dog near the river bank."
tokens = word_tokenize(sentence)
tags = pos_tag(tokens)

# Simple NP: (Determiner) + (Adjective(s)) + Noun
grammar = r"""
  NP: {<DT>?<JJ.*>*<NN.*>+}
"""
chunker = RegexpParser(grammar)
tree = chunker.parse(tags)
print(tree)

# Visualize in a notebook (optional if GUI available)
# tree.draw()  # comment if running on a headless server
```

```python
from nltk import sent_tokenize, word_tokenize, pos_tag
from nltk.chunk import RegexpParser


def pos_and_np_chunks(text):
    sents = sent_tokenize(text)
    grammar = r"NP: {<DT>?<JJ.*>*<NN.*>+}"
    chunker = RegexpParser(grammar)

    outputs = []
    for s in sents:
        tokens = word_tokenize(s)
        tags = pos_tag(tokens)
        chunks = chunker.parse(tags)
        outputs.append((s, tags, chunks))
    return outputs


news = "The central bank raised interest rates to combat rising inflation."
casual = "OMG this new phone is insanely good, totally loving it!"

for label, sample in [("NEWS", news), ("CASUAL", casual)]:
    print(f"\n--- {label} ---")
    for s, tags, chunks in pos_and_np_chunks(sample):
        print("Sentence:", s)
        print("Tags:", tags)
        print("Chunks:", chunks)
```

```python
import nltk
from nltk.corpus import treebank

# Load tagged sentences
tagged_sents = treebank.tagged_sents()
print("Example tagged sentence:", tagged_sents[0])
print("Total sentences:", len(tagged_sents))
```

```python
# 90% train, 10% test
size = int(len(tagged_sents) * 0.9)
train_sents, test_sents = tagged_sents[:size], tagged_sents[size:]
```

```python
from nltk.tag import UnigramTagger

unigram_tagger = UnigramTagger(train_sents)
print("Unigram Tagger Accuracy:", unigram_tagger.evaluate(test_sents))
```

```python
from nltk.tag import DefaultTagger

default_tagger = DefaultTagger("NN")
unigram_backoff = UnigramTagger(train_sents, backoff=default_tagger)
print("Unigram+Default Accuracy:", unigram_backoff.evaluate(test_sents))
```

```python
from nltk.tag import BigramTagger

bigram_tagger = BigramTagger(train_sents, backoff=unigram_backoff)
print("Bigram+Unigram+Default Accuracy:", bigram_tagger.evaluate(test_sents))
```

```python
from nltk.tag import RegexpTagger, UnigramTagger, BigramTagger

# Regex rules
patterns = [(r".*ing$", "VBG"), (r".*ed$", "VBD"), (r".*s$", "NNS"), (r"[0-9]+", "CD")]
regex_tagger = RegexpTagger(patterns)

# Backoff hierarchy
combined_tagger = BigramTagger(
    train_sents, backoff=UnigramTagger(train_sents, backoff=regex_tagger)
)

print("Combined Tagger Accuracy:", combined_tagger.evaluate(test_sents))
```

```python
# Compare predicted vs actual tags
test_sentence = [w for w, t in test_sents[0]]
print("Sentence:", test_sentence)
print("Predicted:", combined_tagger.tag(test_sentence))
print("Actual:", test_sents[0])
```
