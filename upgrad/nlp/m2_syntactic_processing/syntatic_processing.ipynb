```python
# Tokenizers & taggers (handles new/old names)
for pkg in [
    "punkt",
    "punkt_tab",
    "averaged_perceptron_tagger_eng",
    "averaged_perceptron_tagger",
]:
    try:
        nltk.data.find(
            {
                "punkt": "tokenizers/punkt",
                "punkt_tab": "tokenizers/punkt_tab/english",
                "averaged_perceptron_tagger_eng": "taggers/averaged_perceptron_tagger_eng",
                "averaged_perceptron_tagger": "taggers/averaged_perceptron_tagger",
            }[pkg]
        )
    except LookupError:
        nltk.download(pkg, quiet=True)  # Corpora
    for corp in ["brown", "treebank", "gutenberg", "conll2000"]:
        try:
            nltk.data.find(f"corpora/{corp}")
        except LookupError:
            nltk.download(corp, quiet=True)
```

```python
import nltk
from nltk import sent_tokenize, word_tokenize, pos_tag

text = """Dr. Rahul said, "NLP is fun!" Let's build a POS tagger for real-world text like emails, tweets, and news."""  # 1)
# sentences
sents = sent_tokenize(text)
print("Sentences:", sents)  # 2) tokens per sentence
tokenized = [word_tokenize(s) for s in sents]
print("Tokens:", tokenized)  # 3) POS tags per sentence
tagged = [pos_tag(tokens) for tokens in tokenized]
for i, sent_tags in enumerate(tagged, 1):
    print(f"Tagged sentence {i}:", sent_tags)
```

```python
import nltk
from nltk import pos_tag, word_tokenize
from nltk.chunk import RegexpParser
```

```python
sentence = "The quick brown fox jumps over the lazy dog near the river bank."
tokens = word_tokenize(sentence)
tags = pos_tag(tokens)  # Simple NP: (Determiner) + (Adjective(s)) + Noun
grammar = r""" NP: {<DT>?<JJ.*>*<NN.*>+} """
chunker = RegexpParser(grammar)
tree = chunker.parse(tags)
print(tree)
```
