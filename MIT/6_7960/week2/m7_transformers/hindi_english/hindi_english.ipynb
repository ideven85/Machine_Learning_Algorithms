```python
from indicnlp.tokenize.indic_tokenize import trivial_tokenize
from indicnlp.tokenize.sentence_tokenize import sentence_split
from indicnlp.transliterate.unicode_transliterate import UnicodeIndicTransliterator
```

```python
help(trivial_tokenize)
```

```python
text = "I am a geek. I love Natural language processing."
tokens = " ".join(trivial_tokenize(text=text, lang="en"))
print(tokens)
sentences = sentence_split(tokens, lang="en")
print(sentences)
```

```python
converted_text = UnicodeIndicTransliterator.transliterate(
    text, lang1_code="en", lang2_code="hi"
)
print(converted_text)
```

```python
# Kannada Text
text = " एक बेहतरीन प्लेटफार्म है।"
tokens = " ".join(trivial_tokenize(text, lang="hi"))
# Sentence Splitting
sentences = " ".join(sentence_split(tokens, lang="hi"))

print("Original Text:", text)
print("Sentences:", sentences)
converted_text = UnicodeIndicTransliterator.transliterate(sentences, "hi", "pa")
print(converted_text)
```

```python
text = "ఇది ప్రముఖ కంప్యూటర్ సైన్స్ ప్లాట్‌ఫారమ్."

# Script Conversion (Telugu to Hindi)
converted_text = UnicodeIndicTransliterator.transliterate(text, "te", "hi")

print("Original Text:", text)
print("Converted Text:", converted_text)
```
