```python
import numpy as np
import pandas as pd
import sklearn
import requests
import pickle
```

```python
response = requests.get(
    "https://github.com/greydanus/mnist1d/raw/master/mnist1d_data.pkl"
)
URL = "https://github.com/greydanus/mnist1d/raw/master/mnist1d_data.pkl"
```

```python
if response.status_code == 200:
    pickle.dump(response.content, open("mnist1d_data.pkl", "wb"))
```

```python
from urllib.request import urlopen

with open("mnist1d_data.pkl", "r") as f:
    data = pickle.load(urlopen(URL))
```
