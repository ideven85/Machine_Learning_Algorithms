## computation graph in torch

```python
import torch
```

```python
def compute_fn(a, b, c):
    node1 = torch.sub(a, b)
    node2 = torch.mul(node1, 2)
    node3 = torch.add(node2, c)
    return node3
```

```python
print("Scalar Inputs:", compute_fn(torch.tensor(1), torch.tensor(2), torch.tensor(3)))
```

```python
a = torch.tensor(3.0, requires_grad=True)
```

```python
b = a * 2
b.backward()
print(a.grad)
```

## basic nn

```python
w = torch.tensor(1.0, requires_grad=True)
b = torch.tensor(0.5, requires_grad=True)
x = torch.tensor([1.4])
y = torch.tensor([2.1])
```

```python
z = torch.mul(x, w) + b
z
```

```python
loss = (y - (x * w + b)) ** 2
```

```python
loss.sum()
```

```python
loss.backward()
```

```python
print(f"dl/dw: {w.grad}, dl/db: {b.grad}")
```

```python
# verifying
print(2 * x * ((w * x + b) - y))
```
