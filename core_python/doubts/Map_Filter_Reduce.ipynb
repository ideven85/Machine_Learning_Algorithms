```python
import functools

from core_python.decorators.ClockDeco import clock
```

```python
@functools.cache
@clock
def fib(n):
    return n if n <= 1 else fib(n - 1) + fib(n - 2)
```

```python
print(fib(10))
```

```python
print(fib(20))
```

```python
print(fib(25))
```
